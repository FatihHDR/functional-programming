```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

galeri_latihan = []
```

```python
img_original = Image.open(r"D:\Prak. Fungsional\modul6\images\patrick.jpg")

img_original = img_original.convert('RGB')

img_array = np.array(img_original)

print(f"Ukuran gambar asli: {img_array.shape}")
print(f"Tipe data: {img_array.dtype}")

plt.figure(figsize=(8, 6))
plt.imshow(img_array)
plt.title("Gambar Asli")
plt.axis('off')
plt.show()

galeri_latihan.append(("Latihan 0: Original", img_array))
print("Latihan 0: Gambar asli berhasil dimuat!")
```

```python
def buat_penampil(cols=3, figsize=(12,4)): 
    def tampilkan(img_list, titles=None, cmaps=None): 
        rows = (len(img_list) + cols - 1) // cols 
        plt.figure(figsize=figsize) 
 
        for i, img in enumerate(img_list, 1): 
            plt.subplot(rows, cols, i) 
 
            if cmaps and i-1 < len(cmaps) and cmaps[i-1]: 
                plt.imshow(img, cmap=cmaps[i-1]) 
            else: 
                plt.imshow(img) 
 
            plt.axis('off') 
 
            if titles and i-1 < len(titles): 
                plt.title(titles[i-1]) 
 
        plt.tight_layout() 
        plt.show() 
    return tampilkan 
 
 
def tampil_foto(img, posisi=(1,1,1), title=None, cmap=None): 
    plt.subplot(*posisi) 
    if cmap: 
        plt.imshow(img, cmap=cmap) 
    else: 
        plt.imshow(img) 
    if title: 
        plt.title(title) 
        plt.axis("off") 
```

Pada gambar yang sudah kalian upload pada latihan 0, buatlah versi gambar yang diperkecil 
menjadi ukuran 200x200 piksel. Simpan hasilnya ke variabel 'lat1_img'

```python
# Latihan 1: Resize gambar menjadi 200x200 piksel
img_pil = Image.fromarray(img_array)
img_resized = img_pil.resize((200, 200))
lat1_img = np.array(img_resized)

tampil = buat_penampil(cols=2, figsize=(10, 5))
tampil([img_array, lat1_img], 
    titles=[f"Original {img_array.shape[1]}x{img_array.shape[0]}", 
            f"Resized {lat1_img.shape[1]}x{lat1_img.shape[0]}"])

galeri_latihan.append(("Latihan 1: Resize", lat1_img))
print("Latihan 1 tersimpan!")
```

Pada gambar yang sudah kalian upload pada latihan 0, gunakan teknik Lambda/HOF untuk 
meningkatkan kecerahan gambar sebesar 2x lipat. Simpan hasilnya ke variabel 'lat2_img'

```python
# Latihan 2: Meningkatkan kecerahan gambar 2x lipat menggunakan Lambda/HOF
brightness_factor = 2.0
increase_brightness = lambda img: np.clip(img * brightness_factor, 0, 255).astype(np.uint8)
lat2_img = increase_brightness(img_array)

tampil = buat_penampil(cols=2, figsize=(10, 5))
tampil([img_array, lat2_img], 
       titles=["Original", f"Brightness x{brightness_factor}"])

galeri_latihan.append(("Latihan 2: Brightness", lat2_img))
print("Latihan 2 tersimpan!")
```

Pada  gambar  yang sudah kalian upload pada latihan 0, lakukan cropping pada posisi yang 
kalian  inginkan.  Lalu  tampilkan  hasilnya  seperti  contoh  berikut  (tambahkan  sebuah 
kotak/rectangle  untuk  menandai  posisi  yang  dicrop)!  Simpan  hasil  crop  kedalam  variabel 
'lat3_img

```python
# Latihan 3: Cropping gambar dengan menandai area crop menggunakan kotak
height, width = img_array.shape[:2]
x_start, y_start = width // 4, height // 4
x_end, y_end = 3 * width // 4, 3 * height // 4

lat3_img = img_array[y_start:y_end, x_start:x_end]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(img_array)
axes[0].set_title("Original dengan Box")
axes[0].axis('off')

rect = patches.Rectangle((x_start, y_start), x_end - x_start, y_end - y_start,
                        linewidth=3, edgecolor='red', facecolor='none')
axes[0].add_patch(rect)

axes[1].imshow(lat3_img)
axes[1].set_title(f"Cropped {lat3_img.shape[1]}x{lat3_img.shape[0]}")
axes[1].axis('off')

plt.tight_layout()
plt.show()

galeri_latihan.append(("Latihan 3: Box Resize", lat3_img))
print("Latihan 3 tersimpan!")
```

Pada gambar yang sudah kalian upload pada latihan 0, putar gambar asli sebesar 45 derajat. 
Simpan hasilnya ke variabel 'lat4_img'.

```python
# Latihan 4: Rotasi gambar sebesar 45 derajat
img_pil = Image.fromarray(img_array)
img_rotated = img_pil.rotate(45, expand=True, fillcolor=(255, 255, 255))
lat4_img = np.array(img_rotated)

tampil = buat_penampil(cols=2, figsize=(10, 5))
tampil([img_array, lat4_img], 
       titles=["Original", "Rotasi 45"])

galeri_latihan.append(("Latihan 4: Rotasi", lat4_img))
print("Latihan 4 tersimpan!")
```

```python
# Tampilkan semua hasil latihan dalam satu figure
print(f"Total latihan yang berhasil: {len(galeri_latihan)}")
print("="*50)

tampil_galeri = buat_penampil(cols=3, figsize=(15, 10))
images = [img for _, img in galeri_latihan]
titles = [title for title, _ in galeri_latihan]

tampil_galeri(images, titles=titles)

print("\nSemua latihan selesai!")
```
