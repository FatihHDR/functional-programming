```python
import matplotlib.pyplot as plt
from functools import reduce

# Menyiapkan nilai turunan menggunakan list comprehensions / lambdas
jumlah_item = list(map(lambda t: t['jumlah_item'], transaksi))
total_belanja = list(map(lambda t: t['jumlah_item'] * t['harga_per_item'], transaksi))
harga_per_item = list(map(lambda t: t['harga_per_item'], transaksi))

# Buat distribusi status
status_list = list(map(lambda t: t['status'], transaksi))
status_counts = {s: status_list.count(s) for s in set(status_list)}

# Scatter: jumlah_item vs total_belanja
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
colors = ['#1f77b4' if s == 'Lunas' else '#ff7f0e' if s == 'Belum Lunas' else '#d62728' for s in status_list]
sizes = [max(40, min(tb/5000, 400)) for tb in total_belanja]  # visual size scale
plt.scatter(jumlah_item, total_belanja, c=colors, s=sizes, alpha=0.8, edgecolors='k')
for i, t in enumerate(transaksi):
    plt.text(jumlah_item[i], total_belanja[i], t['id'], fontsize=9, ha='center', va='center')
plt.xlabel('Jumlah Item')
plt.ylabel('Total Belanja (Rp)')
plt.title('Scatter: Jumlah Item vs Total Belanja')
plt.grid(True, linestyle='--', alpha=0.4)

# Pie: distribusi status
plt.subplot(1, 2, 2)
labels = list(status_counts.keys())
sizes = list(status_counts.values())
explode = [0.1 if lab == 'Batal' else 0.05 if lab == 'Belum Lunas' else 0 for lab in labels]
colors_pie = ['#d62728' if lab == 'Batal' else '#ff7f0e' if lab == 'Belum Lunas' else '#1f77b4' for lab in labels]
patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors_pie, explode=explode, shadow=True)
for txt in texts + autotexts:
    txt.set_fontsize(9)
plt.axis('equal')
plt.title('Distribusi Status Transaksi')

plt.tight_layout()
plt.show()
```
